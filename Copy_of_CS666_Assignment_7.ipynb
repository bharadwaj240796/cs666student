```python
#
# 4) Now, please study the code of the bookmarklet.
#    See https://gist.github.com/haehn/bbec8cf8442adea5bb10c08d9c3cc13f
#
#    TODO: What is the main difference between processing with the bookmarklet
#          versus the kernel mask filtering from above in terms of accessing
#          the pixels?
#

# The bookmarklet approach often works directly on the pixels of an image that is shown on a webpage in terms of accessing pixels.
# It can detect the image element and retrieve its pixel data by using the Document Object Model (DOM) of the webpage. The bookmarklet can then use JavaScript code to modify the pixels.

# Kernel mask filtering, on the other hand, processes images pixel by pixel without being constrained by a particular web page or browser environment. 
# By iterating over each pixel individually and looking at its neighbors in accordance with the kernel mask, it may access the pixels of a picture. 
# The weights to be applied to each nearby pixel are specified in the kernel mask, and the weighted sum that is produced is utilized to calculate the new value of the processed central pixel.
```
